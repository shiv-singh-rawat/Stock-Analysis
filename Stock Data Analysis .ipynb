# Stock Data Analysis 

The aim of this project is to analyze stock data for three companies (NIFTY, ASIANPAINTS, and BAJAJFINSERV) by performing various statistical and visual analyses. This includes calculating percent changes, plotting histograms, computing basic statistics, analyzing cumulative frequencies, creating confusion matrices, and investigating correlations and moving averages.

```python
#importing the lib.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load stock data
df_nifty=pd.read_csv(r'C:\Users\ABC\Downloads\^NSEI.csv')
df_asian=pd.read_csv(r'C:\Users\ABC\Downloads\ASIANPAINT.NS.csv')
df_bajaj=pd.read_csv(r'C:\Users\ABC\Downloads\BAJAJFINSV.NS.csv')
```

## Data overview

def data_overview(df, name):
    print(f"\n{name} Info:")
    df.info()
    print(f"\n{name} Describe:")
    print(df.describe())
data_overview(df_nifty, 'NIFTY')
data_overview(df_asian, 'ASIANPAINTS')
data_overview(df_bajaj, 'BAJAJFINSERV')

### Data Analysis: Calculate Percent Change

```python
# Calculate percent change
df_nifty['percent_change'] = (df_nifty['Close'] - df_nifty['Close'].shift(1)) / df_nifty['Close'].shift(1) * 100
df_asian['percent_change'] = (df_asian['Close'] - df_asian['Close'].shift(1)) / df_asian['Close'].shift(1) * 100
df_bajaj['percent_change'] = (df_bajaj['Close'] - df_bajaj['Close'].shift(1)) / df_bajaj['Close'].shift(1) * 100
```

## Data visulization

```python
# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(df_nifty['percent_change'], bins=30, edgecolor='black', alpha=0.7)
plt.title('NIFTY Percent Change Histogram')
plt.xlabel('Percent Change')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

This histogram shows the distribution of percent changes in the closing price of NIFTY stocks over a given period. The x-axis represents the range of percent changes, while the y-axis shows the frequency of occurrence for each range. Each bar represents a range of percent changes, and the height of the bar indicates how many times that range occurred in the dataset. This plot helps visualize how frequently different percent changes occur in the NIFTY stock prices.

```python
plt.figure(figsize=(10, 6))
plt.hist(df_asian['percent_change'], bins=30, edgecolor='black', alpha=0.7)
plt.title('ASIANPAINTS Percent Change Histogram')
plt.xlabel('Percent Change')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

This histogram illustrates how the percent changes in ASIANPAINTS' closing stock prices are distributed over a specific time frame. The x-axis represents different ranges of percent changes, while the y-axis shows how frequently each range occurs, measured by frequency. Each bar represents a distinct range of percent changes, and the height of each bar reflects the frequency with which that range appears in the dataset. By examining this visualization, we gain insight into the pattern and frequency of percent changes in ASIANPAINTS stock prices.

```python
plt.figure(figsize=(10, 6))
plt.hist(df_bajaj['percent_change'], bins=30, edgecolor='black', alpha=0.7)
plt.title('BAJAJFINSERV Percent Change Histogram')
plt.xlabel('Percent Change')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

This histogram visualizes the distribution of percent changes in BAJAJFINSERV's closing stock prices over a specific period. On the x-axis, different ranges of percent changes are represented, while the y-axis displays the frequency of occurrence for each range. Each bar corresponds to a particular range of percent changes, with the height of the bar indicating how frequently that range occurs in the dataset. By examining this histogram, we can understand the frequency distribution of percent changes in BAJAJFINSERV stock prices and identify any notable patterns or trends.

```python
# Calculate basic statistics
nifty_stats = df_nifty['percent_change'].describe()
asian_stats = df_asian['percent_change'].describe()
bajaj_stats = df_bajaj['percent_change'].describe()

print("NIFTY Statistics:")
print(nifty_stats)
print("\nASIANPAINTS Statistics:")
print(asian_stats)
print("\nBAJAJFINSERV Statistics:")
print(bajaj_stats)
```

```python
# Plot cumulative frequency
plt.figure(figsize=(10, 6))
plt.plot(nifty_stats.index, nifty_stats.values.cumsum() / len(df_nifty), label='NIFTY', marker='o')
plt.plot(asian_stats.index, asian_stats.values.cumsum() / len(df_asian), label='ASIANPAINTS', marker='o')
plt.plot(bajaj_stats.index, bajaj_stats.values.cumsum() / len(df_bajaj), label='BAJAJFINSERV', marker='o')
plt.title('Cumulative Frequency')
plt.xlabel('Statistics')
plt.ylabel('Cumulative Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

This cumulative frequency chart shows how often the stock prices of Asian Paints, Bajaj Finserv, and  NIFTY 50, were at or below a certain price over an unspecified period of time.

The y-axis shows the cumulative frequency, which is the total number of times a stock's price fell at or below a specific value. For example, if the cumulative frequency for Asian Paints is 0.8, it means the stock price was at or below that value 80% of the time.
The x-axis likely represents the stock prices of the three companies, but the labels are missing, so we cannot say for sure what the exact prices are.
By comparing the lines for the three companies, you can see how their stock prices compare to each other. For example, if the cumulative frequency line for Asian Paints is higher than the line for Bajaj Finserv at a certain point, it means that the stock price of Asian Paints was lower than the stock price of Bajaj Finserv for a greater percentage of the time.

```python
# Plot moving averages
plt.figure(figsize=(10, 6))
plt.plot(df_nifty['Close'].rolling(window=20).mean(), label='NIFTY 20-day MA', color='green')  # Changed window size to 20
plt.plot(df_nifty['Close'].rolling(window=100).mean(), label='NIFTY 100-day MA', color='red')  # Changed window size to 100
plt.title('NIFTY Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

20 MA is above the 100 MA, like in the graph you likely saw, it's generally interpreted as a bullish signal. This suggests a possible upward trend because the short-term price movement (reflected by the 20 MA) is stronger than the long-term trend (reflected by the 100 MA).

```python
plt.figure(figsize=(10, 6))
plt.plot(df_asian['Close'].rolling(window=50).mean(), label='ASIANPAINTS 50-day MA', color='green')
plt.plot(df_asian['Close'].rolling(window=200).mean(), label='ASIANPAINTS 200-day MA', color='red')
plt.title('ASIANPAINTS Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
